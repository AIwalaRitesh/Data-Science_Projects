## Data types of all the attributes:
- Show_id: Numeric identifier (integer)
- Type: Identifier for movie or TV show (object/string)
- Title: Title of the movie or TV show (object/string)
- Director: Director of the movie or TV show (object/string)
- Cast: Actors involved in the movie or TV show (object/string)
- Country: Country where the movie or TV show was produced (object/string)
- Date_added: Date when the movie or TV show was added on Netflix (object/string)
- Release_year: Actual release year of the movie or TV show (integer)
- Rating: TV rating of the movie or TV show (object/string)
- Duration: Total duration of the movie in minutes or number of seasons (object/string)
- Listed_in: Genre of the movie or TV show (object/string)
- Description: Summary description of the movie or TV show (object/string)

## Conversion of categorical attributes to 'category':

## Let's try and find out which are the major countries of content production for the given dataset

```python
(df.groupby("country")["show_id"]
 .count()
 .reset_index()
 .rename(columns = {"show_id":"count"})
 .sort_values("count",ascending = False))
 .head(25)
```

Performing a non-graphical analysis on the value counts and unique attributes of the "country" column, we can observe the following:

1. The country with the highest count of shows/movies available on Netflix is the United States, with 2818 entries.
2. India is the second highest with 972 entries, indicating a significant presence of Indian content on the platform.
3. The United Kingdom has 419 entries, followed by Japan with 245 entries.
4. South Korea and Canada have 199 and 181 entries, respectively.
5. There are multiple countries with a considerable number of entries, such as Spain (145), France (124), Mexico (110), Egypt (106), and Turkey (105).
6. Some entries represent a combination of countries, such as "United Kingdom, United States" (75) and "United States, Canada" (73), indicating co-productions or content available in multiple countries.
7. The dataset includes shows/movies from various regions, including Germany (67), China (66), Argentina (56), Hong Kong (53), Thailand (61), and more.
8. It's important to note that this analysis is based on the available data in the dataset and may not represent the complete Netflix library or reflect the current state of content availability.

Value counts and unique attributes provide insights into the distribution and diversity of content based on the countries listed in the dataset.

```python
# Group the data and calculate the count
grouped_data = df.groupby(["rating"])["show_id"].count().reset_index().rename(columns = {"show_id":"count"})

# Sort the data by count in descending order
sorted_data = grouped_data.sort_values("count", ascending=False)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sorted_data["count"], labels=sorted_data["rating"], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Distribution of Ratings")
plt.show()
```

```python
(df.groupby(["rating","country"])["show_id"]
 .count()
 .reset_index()
 .rename(columns = {"show_id":"count"})
 .sort_values("count",ascending = False)).head(25)
```

```python
# Group the data and calculate the count
grouped_data = df.groupby(["rating", "country"])["show_id"].count().reset_index().rename(columns = {"show_id":"count"})

# Sort the data by count in descending order
sorted_data = grouped_data.sort_values("count", ascending=False).head(25)

# Create the grouped bar plot
plt.figure(figsize=(12, 8))
x_ticks = np.arange(len(sorted_data))
plt.bar(x_ticks, sorted_data["count"], align='center', alpha=0.8)
plt.xticks(x_ticks, sorted_data["rating"] + " - " + sorted_data["country"], rotation=90)
plt.xlabel("Rating - Country")
plt.ylabel("Count")
plt.title("Count Distribution of Ratings and Countries")
plt.tight_layout()
plt.show()
```
